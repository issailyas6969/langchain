```python
from langchain_community.document_loaders import TextLoader
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
from langchain_community.document_loaders import WikipediaLoader
docs=WikipediaLoader(query="Adolf Hitler").load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docss=splitter.split_documents(docs)
docss
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embedding=OllamaEmbeddings(model="gemma:2b")
from langchain_community.vectorstores import FAISS
vector_db=FAISS.from_documents(docss,embedding)
vector_db
```

```python
from langchain_community.llms import Ollama
llm=Ollama(model="gemma:2b")
```

```python
from langchain.chains.combine_documents import StuffDocumentsChain

prompt=ChatPromptTemplate.from_messages(
    """
answer the following question based on context:
<context>
{context}
</context>

    """
)

document_chain = StuffDocumentsChain(
    llm,
    prompt
)
document_chain
```

```python
document_chain.invoke({
    "input":"",
    "context":[Document(page_content="")
})
```

```python
retriever=vector_db.as_retriever()
retriever
```

```python
from langchain_chains import create_retrieval_chains
retrival_chain=create_retrieval_chains(retriever,document_chain)
```

```python
response=retrival_chain.invoke({"input":""})
response["answer"]
```
